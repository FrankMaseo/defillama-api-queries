```python
import requests as re

API_URL = 'http://api.llama.fi/'

headers = {
    'Accept': '*/*',
    'User-Agent': """Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"""
}

def defillama(endpoint, params={}, json={}):
    resp = re.get(
        API_URL + endpoint,
        headers = headers
    )
    return resp.text

def bridge(endpoint, params = {}):
    resp = re.get(
        'https://bridges.llama.fi/' + endpoint,
        headers = headers,
        params = params
    )
    
    return resp.text
```

```python
import json

protocols = json.loads(defillama(
    'protocols'
))
```

```python
bridges = []

for p in protocols:
    if 'BRIDGE' in p['category'].upper() or 'CROSS CHAIN' in p['category'].upper():
        bridges.append(p)
```
